```python
import streamlit as st
import joblib
import numpy as np

st.title("ü©∫ Diabetes Prediction App")

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÖŸàÿØŸäŸÑ ŸàÿßŸÑÿ≥ŸÉÿßŸäŸÑÿ±
model = joblib.load("saved_models/best_rf_model.joblib")
scaler = joblib.load("saved_models/scaler.joblib")

st.write("### ÿ£ÿØÿÆŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÖÿ±Ÿäÿ∂ ŸÑŸÑÿ™ŸÜÿ®ÿ§ ÿ®ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿßŸÑÿ•ÿµÿßÿ®ÿ© ÿ®ÿßŸÑÿ≥ŸÉÿ±Ÿä:")

# ‚úÖ ŸÜŸÅÿ≥ ÿ™ÿ±ÿ™Ÿäÿ® ÿßŸÑÿ™ÿØÿ±Ÿäÿ® (8 features)
age = st.number_input("Age", min_value=1, max_value=120)
bmi = st.number_input("BMI", min_value=10.0, max_value=70.0)
systolic_bp = st.number_input("Systolic BP", min_value=50.0, max_value=250.0)
diastolic_bp = st.number_input("Diastolic BP", min_value=30.0, max_value=150.0)
glucose = st.number_input("Glucose", min_value=50.0, max_value=400.0)
hdl = st.number_input("HDL", min_value=10.0, max_value=100.0)
triglyceride = st.number_input("Triglyceride", min_value=20.0, max_value=600.0)
gender = st.selectbox("Gender", ["Female", "Male"])
gender_male = 1 if gender == "Male" else 0

if st.button("üîç Predict Diabetes Risk"):
    features = np.array([[age, bmi, systolic_bp, diastolic_bp, glucose, hdl, triglyceride, gender_male]])
    features_scaled = scaler.transform(features)
    probability = model.predict_proba(features_scaled)[0, 1]
    st.success(f"üîπ Estimated Probability of Having Diabetes: **{probability * 100:.2f}%**")

    if probability > 0.5:
        st.error("‚ö†Ô∏è High risk of diabetes ‚Äî consult a doctor.")
    else:
        st.info("‚úÖ Low risk of diabetes.")
```
